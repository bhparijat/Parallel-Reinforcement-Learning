```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
env = gym.envs.make('FrozenLake8x8-v0')
```

```python
q_table = np.random.randn(64,env.action_space.n)
```

```python
learning_rate = 0.1
discount_factor = 0.999 
max_step = 100
```

```python
rewards = []
import random
for i in tqdm(range(300000)):
    #print(i)
    state = env.reset()
    steps = 0
    beta = discount_factor
    terminal = False
    total_rewards = 0
    while terminal==False:
        
        action =  q_table[state].argmax() if random.random()>0.3 else np.random.randint(4,size=1)[0]
        next_state,reward,terminal,_= env.step(action)
        
        q_table[state][action] = q_table[state][action] + learning_rate*(reward*1000 + beta*q_table[next_state].max()-q_table[state][action])
        
        state = next_state
        beta*=discount_factor
        total_rewards+=reward
        steps+=1
    if i%100 == 0:
        #print(total_rewards,reward)
        rewards.append(total_rewards)
```
